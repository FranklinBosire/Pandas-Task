```python
import pandas as pd
from matplotlib import style
style.use('ggplot')
data = pd.read_excel('/content/drive/MyDrive/pandas/File.xlsx')
```

```python
#display (pd.DataFrame(data)) 

data.head()
```

```python
import matplotlib.pyplot as plt
data.describe()

```

```python
plt.matshow(data.corr())
continuous_features = data.describe().columns
plt.xticks(range(len(continuous_features)), continuous_features, rotation="50")
plt.yticks(range(len(continuous_features)), continuous_features, rotation="50")
plt.colorbar()
plt.show()
```

```python
# 1.Is it more expensive to live in FL or NY?
import matplotlib.pyplot as plt
data[["Location", "Price"]].groupby("Location").mean()


```

```python
# 2. is the crime rate higher in FL or NY? 
data[["Location", "Crime Rating"]].groupby("Location").mean()
```

```python
# 3.Is the crime rate higher in lower or higher house price areas?

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.scatterplot(x='Price',y='Crime Rating',data= data)
```
